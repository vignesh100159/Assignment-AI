```python
corpus = df1["reviews.text"].tolist()
```

```python
from nltk.corpus import stopwords
from string import punctuation
```

```python
remove = list(stopwords.words("english"))+list(punctuation)
```

```python
from nltk.tokenize import word_tokenize
from nltk.stem.lancaster import LancasterStemmer
lancaster_stemm = LancasterStemmer()
stemm = LancasterStemmer()
```

```python
r_corpus= []
for i in range(len(corpus)):
    text = word_tokenize(corpus[i].lower())
    text = [lancaster_stemm.stem(y) for y in text if y not in remove]
    sent = " ".join(text)
    
    r_corpus.append(sent)
```

```python
df2 = pd.DataFrame(r_corpus)
```

```python
df2["reviews.rating"] = df1["reviews.rating"]
```

```python
df2.head()
```

```python
df2.columns = ["reviews.text", "reviews.rating"]
```

```python
df2.head()
```

```python
r_corpus2= []
for i in range(len(r_corpus)):
    text = word_tokenize(r_corpus[i])
    tag = nltk.pos_tag(text)
    adj_tags = ["JJ", "JJR", "JJS"]
    close_adjectives = [a[0] for a in tag if a[1] in adj_tags]
    sent = " ".join(text)
    r_corpus2.append(sent)
```

```python
list1 = []
for i in range (len(r_corpus)):
    count = dict()
    wordcount = r_corpus[i].split(" ")
    
    for words in wordcount:
        if words in count:
            count[words] += 1
        else:
            count[words] = 1
    list1.append(count)
```
